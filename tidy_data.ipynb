### 3. Cake Recipes

    a. Read the cake_recipes.csv data. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.
    
    b. Tidy the data as necessary.
    
    c. Which recipe, on average, is the best? recipe b
    
    d. Which oven temperature, on average, produces the best results? 275
    
    e. Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

```python
# import csv
cake = pd.read_csv('cake_recipes.csv')
```

```python
# observe df
cake.head()
```

```python
# create melt dataframe with recipe/position as index
cake_melt = cake.melt(id_vars = 'recipe:position')
# rename to remove :
cake_melt.rename(columns = {'recipe:position':'recipe_position'}, inplace = True)
```

```python
# create df for split recipe and position
recipe_and_position_df = cake_melt.recipe_position.str.split(':', expand = True)

#rename resulting columns
recipe_and_position_df.rename(columns = {0:'recipe', 1:'position'}, inplace = True)
```

```python
# create concat'd df for melt and recipe/position
cake2 = pd.concat([cake_melt, recipe_and_position_df], axis = 1).drop(columns='recipe_position')
```

```python
# rename columns for ease of access
cake2.rename(columns = {'variable':'temperature', 'value':'score'}, inplace = True)
```

```python
# observe df
cake2
```

```python
# create tidy df that is a pivot table of previous df
# index is the recipe/position with temperature as the column and the score is the values
cake_tidy = cake2.pivot_table(index = ['recipe','position'], columns = 'temperature', values='score')
```

```python
# observe df
cake_tidy
```

```python
# restructure df so recipe/pos are no longer indices
cake_tidy.columns.name = ''
cake_tidy.reset_index(inplace = True)
cake_tidy
```

### Best recipe on average?

```python
# create groupby mean
recipe_avg = cake_tidy.groupby(cake_tidy.recipe).mean()
# read average across df rows
recipe_avg.mean(axis=1)
```

- B

### Best temperature on average?

```python
# mean of columns
cake_tidy.mean()
```

- 275

### Best combination?

```python
# read max value of each row to find highest score
cake_tidy.max(axis = 1)
```

- Recipe : B, Pos: bottom, Temp: 300
